```python
import pandas as pd
import os
```

### Reading the text files to generate a dataframe

```python
os.chdir(os.getcwd()+"/dataset/Hulth2003/Data")
path = os.getcwd()
```

- Creating a dictionary to store all the values extracted from the text files
- Each dictionary stores a key value pair; for instance title_list has the extracted title of all the documents with its doc number

```python
title_list = {}
abstract_list = {}
contr_keyword_list = {}
uncontr_keyword_list = {}
```

### Function to extract title and abstract from the text file

```python
def extract_title_abstract(data):
    text = []
    counter = 0
    locations = []
    while(counter != len(data)):
        if data[counter].startswith('\t'):
            text.append(data[counter].strip())
        else:
            text.append(data[counter].strip())
            locations.append(counter)
        counter += 1
    title = "".join(text[locations[0]:locations[1]])
    abstract = " ".join(text[locations[1]:])
    
    return title, abstract
```

### Function to extract the controlled and uncontrelled keywords

```python
def extract_keywords(data):
    keywords = []
    temp = "".join(data).split(';')
    for val in temp:
        val = val.strip()
        val = val.replace('\t',' ')
        val = val.replace('\n',' ')
        keywords.append(val)
    return keywords
```

### Iterating over all the text files

```python
print(f'Processing files in : {path}')

for file in os.listdir(path):
    # reading a file
    print(f'Reading : {file}')
    f = open(file, 'r')
    data = f.readlines()
    f.close()
    if 'uncontr' in file:
        keywords_uncontr = extract_keywords(data)
        file = int(file.split('_')[0])
        uncontr_keyword_list[file] = ",".join(keywords_uncontr)
    elif 'abstr' in file:
        title, abstract = extract_title_abstract(data)
        file = int(file.split('_')[0])
        title_list[file] = title
        abstract_list[file] = abstract
    else:
        keywords_contr = extract_keywords(data)
        file = int(file.split('_')[0])
        contr_keyword_list[file] = ",".join(keywords_contr)   
```

### Dataframe with Titles of all the files

```python
title_df = pd.DataFrame(title_list.items(), columns=['Doc_no','Title'])
title_df.sort_values(by=['Doc_no'], inplace=True)
title_df.reset_index(drop=True, inplace=True)
title_df
```

### Dataframe with Abstracts of all the files

```python
abstract_df = pd.DataFrame(abstract_list.items(), columns=['Doc_no','Abstract'])
abstract_df.sort_values(by=['Doc_no'], inplace=True)
abstract_df.reset_index(drop=True, inplace=True)
abstract_df
```

### Dataframe with Controled Keywords of all the files

```python
contr_keywords_df = pd.DataFrame(contr_keyword_list.items(), columns=['Doc_no','Control_keywords'])
contr_keywords_df.sort_values(by=['Doc_no'], inplace=True)
contr_keywords_df.reset_index(drop=True, inplace=True)
contr_keywords_df
```

### Dataframe with Uncontroled Keywords of all the files

```python
uncontr_keywords_df = pd.DataFrame(uncontr_keyword_list.items(), columns=['Doc_no','Uncontrol_keywords'])
uncontr_keywords_df.sort_values(by=['Doc_no'], inplace=True)
uncontr_keywords_df.reset_index(drop=True, inplace=True)
uncontr_keywords_df
```

### Merging the keywords
- The following code block concantes both the controled and uncoltroled keywords in a singel colums keywords

```python
contr_keywords_df['Keywords'] = contr_keywords_df['Control_keywords'] + ',' + uncontr_keywords_df['Uncontrol_keywords']
keywords_df = contr_keywords_df[['Doc_no','Keywords']]
keywords_df
```

### Merging the dataframes title and abstract

```python
title_abs_df = pd.merge(title_df, abstract_df, on="Doc_no")
title_abs_df
```

### Generating the final datafram and saving it in a pickle file

```python
data = pd.merge(title_abs_df, keywords_df, on="Doc_no")
data
```

```python
pd.to_pickle(data, '../../../Data/Hulth2003_data.pkl'
```
