```python
import pandas as pd
import os
```

### Reading the text files to generate a dataframe

```python
os.chdir(os.getcwd()+"/SemEval2017")
path = os.getcwd()
```

- Creating a dictionary to store all the values extracted from the text files
- Each dictionary stores the key(document number) and text extracted from the file

```python
text_files = sorted(os.listdir(path+"/doc"))
key_files = sorted(os.listdir(path+"/keys"))
```

### Extracting the texts from the files

```python
text_data = {}
```

```python
for file in text_files:
    print(f'Processing : {file}\n')
    f = open(path+'/doc/'+file,'r',encoding="utf8")
    data = f.read()
    data = data.replace('\n',' ')
    file = (file.split('_')[0][:-4])
    text_data[file] = data
    f.close()
```

```python
text_data
```

### Extracting the keywords from the files

```python
keywords_data = {}
```

```python
for file in key_files:
    print(f'Processing : {file}\n')
    temp = []
    f = open(path+'/keys/'+file,'r',encoding="utf8")
    for line in f.readlines():
        temp.append(line.strip())
    file = (file.split('_')[0][:-4])
    keywords_data[file] = ",".join(temp)
    f.close()
```

```python
keywords_data
```

### Generating dataframe for the texts

```python
texts_df = pd.DataFrame(text_data.items(), columns=['Doc_no','Text'])
texts_df.sort_values(by=['Doc_no'], inplace=True)
texts_df.reset_index(drop=True, inplace=True)
texts_df
```

### Generating dataframe for the keys

```python
keys_df = pd.DataFrame(keywords_data.items(), columns=['Doc_no','Keywords'])
keys_df.sort_values(by=['Doc_no'], inplace=True)
keys_df.reset_index(drop=True, inplace=True)
keys_df
```

### Merging two dataframes

```python
data = pd.merge(texts_df, keys_df, on="Doc_no")
data
```

```python
from rouge_score import rouge_scorer
import numpy as np
```

```python
"""
This function returns the average of precision, recall and f1score
based on the score from rouge1, rouge2, rougeL rounding off the final 
value to 2 decimal places for each doc in the dataset
"""
def get_avg_scores(result):
    precision_score = []
    recall_score = []
    f1_score = []
    for rouge in ['rouge1','rouge2','rougeL']:
        precision_score.append(result[rouge][0])
        recall_score.append(result[rouge][1])
        f1_score.append(result[rouge][2])
    precision_score = np.round(np.average(precision_score), 2)
    recall_score = np.round(np.average(recall_score), 2)
    f1_score = np.round(np.average(f1_score), 2)
    return precision_score, recall_score, f1_score
```

```python
def get_Rouge_scores(refrence_data, test_data):
    precision_score = []
    recall_score = []
    f1_score = []
    Doc_no = data['Doc_no'].tolist()
    Text = data['Text'].tolist()
    
    for refrence, test in zip(refrence_data, test_data):
        
        # modifying the format as a string with spaces
        refrence = refrence.replace(",", " ")
        test = test.replace(",", " ")
        scorer = rouge_scorer.RougeScorer(['rouge1','rouge2','rougeL'], use_stemmer=True)
        scores = scorer.score(refrence, test)
        
        # Getting the avg scores and appending them to the list
        p_score, r_score, f_score = get_avg_scores(scores)
        precision_score.append(p_score)
        recall_score.append(r_score)
        f1_score.append(f_score)
        
    results_df = pd.DataFrame(zip(Doc_no,Text,precision_score, recall_score, f1_score), columns=['Doc_no', 'Text','Avg_precision_score','Avg_recall_score','Avg_f1_score'])
    return results_df
```

```python
"""
This method returns the average scores for 
Precison, Recall and F1 
on the whole document
"""
def get_final_results(result):
    return np.average(result['Avg_precision_score']),np.average(result['Avg_recall_score']),np.average(result['Avg_f1_score'])
```

```python
def plot_wordcloud(text,title = 'Most common top  keyphrases in the articles'):
    
    wordcloud = WordCloud(min_font_size=12,max_font_size=80 ,width=1200, height=800).generate(text)
    plt.figure(figsize=(10,15))
    
    plt.title(title, fontdict={'size': 10, 'color': 'green', 
                                      'verticalalignment': 'bottom'})
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
import pke
from pke.unsupervised import TopicRank

def get_topicrank_keywords():    
    
    text_key = []
    kw=[]

    for text in data['Text']:
        extractor = TopicRank()
        extractor.load_document(input=text,
                        language="en",
                        normalization='stemming')
        
        extractor.candidate_selection(pos={'NOUN', 'PROPN', 'ADJ'})
        extractor.candidate_weighting(threshold=0.74,
                              method='average')
        
        keyphrase = extractor.get_n_best(n=10, stemming=True)
        print(keyphrase)
       
    
    data['Pytopicrank_keyword'] = keyphrase  
```
